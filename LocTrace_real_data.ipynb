Hypothesentests

What effect does a meaningful presentation of location data have on the privacy concern? 

```python
#https://machinelearningmastery.com/statistical-hypothesis-tests-in-python-cheat-sheet/  

from scipy.stats import shapiro
from scipy.stats import ttest_rel
from scipy.stats import wilcoxon


print("\n")  
print("PC Overall Mean1:" , pc_mean)
print("PC Control Mean1:", pc_ctrl_mean)
print("PC Awareness Mean1:", pc_awa_mean)
print("PC Collection Mean1:", pc_coll_mean)
print("\n")

print("\n")
print("PC Overall Mean2:", pc_mean_2)
print("PC Control Mean2:", pc_ctrl_mean_2)
print("PC Awareness Mean2:", pc_awa_mean_2)
print("PC Collection Mean2:", pc_coll_mean_2)
print("\n")

#Funktionen für Hypothesentests
#T-Test für normalverteilte werte
def t_testing (x1,x2):   
    stat, p = ttest_rel(x1, x2)
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
    	print('H0 wird angenommen\n')
    else:
    	print('H0 wird abgelehnt\n')
         
#wilcoxon, wenn normalverteilung nicht vorliegt
def wilcoxon_testing (x1,x2):
    stat, p = wilcoxon(x1, x2)
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
    	print('H0 wird angenommen\n')
    else:
    	print('H0 wird abgelehnt\n')

print('H0: Visualisation of location data does not influence PC\n')
#Shapiro-Wilk Normality Test 
stat0, p0 = shapiro(pc_mean)
stat1, p1 = shapiro(pc_mean_2)

stat2, p2 = shapiro(pc_ctrl_mean)
stat3, p3 = shapiro(pc_ctrl_mean_2)

stat4, p4 = shapiro(pc_awa_mean)
stat5, p5 = shapiro(pc_awa_mean_2)

stat6, p6 = shapiro(pc_coll_mean)
stat7, p7 = shapiro(pc_coll_mean_2)

print('stat0=%.3f, p0=%.3f' % (stat0, p0))
print('stat1=%.3f, p1=%.3f' % (stat1, p1))
if p0 > 0.05 and p1 > 0.05:
 	# do Dependent T-Test
    print("T TEST PC OVERALL MEANS")
    t_testing(pc_mean,pc_mean_2)
else:
 	# Wilcoxon 
    print("WILCOXON PC OVERALL MEANS")
    wilcoxon_testing(pc_mean,pc_mean_2)
    
print('stat2=%.3f, p2=%.3f' % (stat2, p2))
print('stat3=%.3f, p3=%.3f' % (stat3, p3))
if p2 > 0.05 and p3 > 0.05:
 	# do Dependent T-Test
    print("T TEST PC CONTROL MEANS")
    t_testing(pc_ctrl_mean,pc_ctrl_mean_2)
else:
 	# Wilcoxon 
    print("WILCOXON PC CONTROL MEANS")
    wilcoxon_testing(pc_ctrl_mean,pc_ctrl_mean_2)
    
print('stat4=%.3f, p4=%.3f' % (stat4, p4))
print('stat5=%.3f, p5=%.3f' % (stat5, p5))
if p4 > 0.05 and p5 > 0.05:
 	# do Dependent T-Test
    print("T TEST PC AWARENESS MEANS")
    t_testing(pc_awa_mean,pc_awa_mean_2)
else:
 	# Wilcoxon 
    print("WILCOXON PC AWARENESS MEANS")
    wilcoxon_testing(pc_awa_mean,pc_awa_mean_2)
    
print('stat6=%.3f, p6=%.3f' % (stat6, p6))
print('stat7=%.3f, p7=%.3f' % (stat7, p7))
if p6 > 0.05 and p7 > 0.05:
 	# do Dependent T-Test
    print("T TEST PC COLLECTION MEANS")
    t_testing(pc_coll_mean,pc_coll_mean_2)
else:
 	# Wilcoxon 
    print("WILCOXON PC COLLECTION MEANS")
    wilcoxon_testing(pc_coll_mean,pc_coll_mean_2)
```

Alter, Geschlecht und IT Bereich WTP?

Alter weglassen, da eine Stichprobe=1 ist 

```python
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu

#T-Test wenn Normalverteilung vorhanden
def t_testing_ind (x1,x2): 
  stat, p = ttest_ind(x1,x2)
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
	  print('H0 wird angenommen')
  else:
	  print('H0 wird abgelehnt')

#Mann-Whitney U Test wenn keine Normalverteilung
def mannwu_test (x1,x2):
  stat, p = mannwhitneyu(x1, x2)
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
	  print('H0 wird angenommen')
  else:
	  print('H0 wird abgelehnt')

#male and female wtp means
m_wtp_mean = [0.496667,0.463333,0.463333,0.463333,0.463333,0.496667,0.463333,0.496667,0.6,0.496667,0.6,0.496667]
f_wtp_mean = [0.463333,0.6,0.496667]

#IT Job
IT_wtp_mean = [0.463333,0.463333,0.463333,0.496667,0.6]
noIT_wtp_mean = [0.496667,0.463333,0.6,0.463333,0.463333,0.496667,0.496667,0.6,0.496667,0.496667]

#Varianz messen
print(np.var(m_wtp_mean), np.var(f_wtp_mean))
var1 = (np.var(f_wtp_mean))/(np.var(m_wtp_mean))
print(var1)
# Verhältnis der größeren Stichprobenvarianz zur kleineren Stichprobenvarianz <4 somit können wir testen
print(np.var(IT_wtp_mean), np.var(noIT_wtp_mean))
var2 = (np.var(IT_wtp_mean))/(np.var(noIT_wtp_mean))
print(var2)
# Verhältnis der größeren Stichprobenvarianz zur kleineren Stichprobenvarianz <4 somit können wir testen

print('\nH0: Gender has no influence on WTP mean')
stat8, p8 = shapiro(m_wtp_mean)
stat9, p9 = shapiro(f_wtp_mean)
print('stat8=%.3f, p8=%.3f' % (stat8, p8))
print('stat9=%.3f, p9=%.3f' % (stat9, p9))
if p8 > 0.05 and p9 > 0.05:
 	# do Dependent T-Test
    print("T TEST F/M WTP MEANS")
    t_testing_ind(m_wtp_mean,f_wtp_mean)
else:
 	# MannWU Test
    print("MANN WHITNEY U F/M WTP MEANS")
    mannwu_test(m_wtp_mean,f_wtp_mean)

print('\nH0: IT Job has no influence on WTP mean')
stat10, p10 = shapiro(IT_wtp_mean)
stat11, p11 = shapiro(noIT_wtp_mean)
print('stat8=%.3f, p8=%.3f' % (stat10, p10))
print('stat9=%.3f, p9=%.3f' % (stat11, p11))
if p10 > 0.05 and p11 > 0.05:
 	# do independent T-Test
    print("T TEST IT WTP MEANS")
    t_testing_ind(IT_wtp_mean,noIT_wtp_mean)
else:
 	# MannWU Test
    print("MANN WHITNEY U IT WTP MEANS")
    mannwu_test(IT_wtp_mean,noIT_wtp_mean)


```
